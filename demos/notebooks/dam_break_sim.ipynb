```python
# Import inductiva module
import inductiva
```

```python
# [TODO] Give some instructions on how to get the API key
inductiva.api_key = "YOUR_API_KEY"
```

```python
# Define the dam break scenario
scenario = inductiva.fluids.scenarios.DamBreak()
```

```python
# Simulate the scenario with SPlisSPlasH simulator
simulation_output = scenario.simulate(device="gpu",
                                      engine="SPlisHSPlasH",
                                      resolution="low",
                                      simulation_time=1.,
                                      output_dir="outputs")
```

```python
# Render a video of the simulation
# Note: this will only work for SPlisHSPlasH simulations
simulation_output.render(color="blue", alpha=0.8)
```
