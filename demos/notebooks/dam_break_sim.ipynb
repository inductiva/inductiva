## Create you scenario and run a simulation

```python
# Define the dam break scenario
scenario = inductiva.fluids.DamBreak()
```

```python
# Simulate the scenario with a default simulator
simulation_output = scenario.simulate(resolution="low",
                                      simulation_time=1.,
                                      output_dir="outputs")
```

## Render a visualization
