```python
# Import Inductiva module
import inductiva
import os
```

```python
# [TODO] Give some instructions on how to get the API key
inductiva.api_key = "YOUR_API_KEY"
```

```python
# Provide a full path to the input files directory
SIM_DIR = "path/to/input_files"
```

```python
# Initialize a SPlisHSPlasH simulation with the input files
sph_sim = inductiva.fluids.simulators.SPlisHSPlasH()
```

```python
# Run the simulation on the server
output_path = sph_sim.run(
    input_dir=SIM_DIR,
    sim_config_filename="fluid_tank.json",
    device="gpu", 
    output_dir="sim_outputs",
)
```

```python
# Cnvert the simulation output to NetCDF format
inductiva.fluids.post_processing.splishsplash.convert_vtk_data_dir_to_netcdf(os.path.join(output_path, "vtk"), output_time_step=0.01, netcdf_data_dir=os.path.join(output_path, "netcdf")) 
```

```python
# Create a simulation movie 
visualization = inductiva.fluids.SimulationOutput(output_path)
visualization.render()
```
