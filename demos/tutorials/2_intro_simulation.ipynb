# Exploring configuration spaces at scale with Inductiva API: a coastal dynamics use case

Inductiva API allows users to submit previously prepared simulation scripts directly to the simulation package that they are already familiar with. In this way, users can immediatelly take advantage of accelerated hardware to speed up their simulations, without having to change any of their existing simulation scripts. 

Still, most users want to execute more than just one simulation. In fact, users may need to explore the design space of a given simulation scenario to evaluate the impact of varying some design parameters. Inductiva API provides a simple mechanism for helping users perform large scale parameter exploration, starting from a base simulation script. The only thing that is required is to substitute in the simulation script the fixed numerical values of the parameters to explore with user defined tags. This will indicate the API engine that we want to vary these parameters over a range of values defined by the user. This way, users can introduce variations in their simulations with minor effort, and immediately benefit from efficient parallelization strategies to explore the design space of the simulation scenario at scale.

In this tutorial we will show such parameter exploration capabilities on a coastal dynamics scenario, powered by the [SWASH](https://swash.sourceforge.io/download/download.htm) simulation package. SWASH simulates unsteady and free-surface flow and transport phenomena in coastal waters as driven by waves, tides, buoyancy and wind forces. It provides a general basis for describing wave transformations from deep water to a beach, port or harbour, complex changes to rapidly varied flows, and density driven flows in coastal seas, estuaries, lakes and rivers.
