# Inductiva API: Simulation scale made simple, from your laptop

This notebook illustrates how simple it is to run simulations using **Inductiva API**. As you will see, users do not have to deal with the complexity of installing and configuring simulation software packages or of setting up the underlying computatinal infrastructure.

Our API totally abstracts operational issues: users only needs to code a few lines of Python and they can immediately run the simulation on remote hardware. In this way, users can focus on the problem they are trying to solve via simulation instead of wasting time with software and hardware issues that slow down the modeling and experimantion process.


# Fluids Dynamics

In this notebook, we focus on a classic simulation of fluid dynamics: the dam break scenario, where a block of water is let to flow, only subject to gravity and internal forces.

In the figure below, we have a block of water in a side view that is solely defined by its dimensions: ***height, width and length***.


Screenshot 2023-02-23 at 16.37.48.png



For simulating this scenario, we will use a simulation technique called Smoothed Particle Hydrodynamics (SPH), in which the fluid is represented as a set of particles that describe the properties of the fluid. This technique is specially interesting for simulating the interaction between water and the atmosphere, and allows for very realistic simulations.

1. - **install the depedencies**;
2. - **import inductiva packages**;

# Run the DamBreak `scenario`

A first step is to import the `inductiva` module and add your user API key!

Request your personal key by completing this [form](https://docs.google.com/forms/d/e/1FAIpQLSflytIIwzaBE_ZzoRloVm3uTo1OQCH6Cqhw3bhFVnC61s7Wmw/viewform?usp=sf_link). Add your API key below and you can start simulating right away!

```python
import inductiva

# Set the provided API token to be able to access our hardware
inductiva.api_key="YOUR_KEY_HERE"
```

It is time to run the simulation with a few lines of code! We choose a scenario, run the simulation and visualize the results.


```python
dambreak_scenario = inductiva.fluids.DamBreak()

sim_output = dambreak_scenario.simulate()
```
