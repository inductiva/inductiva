### Let's run a simple Fluids simulation

We can do this with very few lines of code!
Recalling the map of the world above these imports, we just need to choose a scenario, run the simulation and visualize the results.


```python
from inductiva import fluids

scenario = fluids.scenarios.DamBreak()
sim_output = scenario.simulate(device="gpu")
video = sim_output.render()

# Show video
video
```
