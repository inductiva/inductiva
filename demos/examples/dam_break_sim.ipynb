```python
# Import inductiva module
import inductiva
```

```python
# Define the dam break scenario
scenario = inductiva.fluids.scenarios.DamBreak(fluid=inductiva.fluids.WATER,
                                               dimensions=(0.3, 1.0, 0.8))
```

```python
# Simulate the scenario with SPlisSPlasH simulator
simulation_output = scenario.simulate(device="cpu",
                                      engine="SPlisHSPlasH",
                                      resolution="medium",
                                      simulation_time=1.,
                                      output_dir="sim_output")
```

```python
# Render a video of the simulation
# Note: this will only work for SPlisHSPlasH simulations
simulation_output.render(color="blue", alpha=0.6)
```
